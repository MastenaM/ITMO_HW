# Посчитайте процент выживаемости у каждого класса пассажиров (Pclass) - 1 балл

```python
total_class1 = len(df[df['Pclass'] == 1])
total_class2 = len(df[df['Pclass'] == 2])
total_class3 = len(df[df['Pclass'] == 3])
survived_class1 = len(df[(df['Pclass'] == 1) & (df['Survived'] == 1)])
survived_class2 = len(df[(df['Pclass'] == 2) & (df['Survived'] == 1)])
survived_class3 = len(df[(df['Pclass'] == 3) & (df['Survived'] == 1)])
rate_surv1 = survived_class1*100/total_class1
rate_surv2 = survived_class2*100/total_class2
rate_surv3 = survived_class3*100/total_class3
print(rate_surv1)
print(rate_surv2)
print(rate_surv3)
```

```python
df[['Surname','Name']] = df['Name'].str.split(",",expand=True)
```

```python
df.head(3) #получили разделение столбцов имени и фамилии
```

```python
#Теперь нам нужно разделить имена на несколько столбцов
```

```python
df['Name'] = df['Name'].str.replace(' ', '', n=1)
df['Name'].head(3)
```

```python
df[['Appeal','Name','Name2','OtherPartofName']] = df['Name'].str.split(' ', expand=True, n=3)
df.head(5)
```

## топ по столбцу Name (по первому имени)
топ мужских имен по первому имени

```python
male_names = df[df['Sex'] == 'male']

male_name1_most = male_names['Name'].mode()[0]
male_name1_most
```

```python
top_male_names1 = male_names['Name'].value_counts(sort=True)
top_male_names1.head(10)
```

топ женских имен по первому имени

```python
female_names = df[df['Sex'] == 'female']
female_name1_most = female_names['Name'].mode()[0]
female_name1_most
```

```python
top_female_names1 = female_names['Name'].value_counts(sort=True)
top_female_names1.head(10)
```

## топ по столбцу Name2 (второе имя).
Для этого первым делом приводим в чистый вид

```python
df['Name2'] = df['Name2'].str.replace('[( )"]', '', regex=True)
df['Name2']
```

топ мужских имен по второму имени

```python
male_name2_most = male_names['Name2'].mode()[0]
male_name2_most
```

```python
top_male_names2 = male_names['Name2'].value_counts(sort=True)
top_male_names2.head(10)
```

топ женских имен по второму имени

```python
female_name2_most = female_names['Name2'].mode()[0]
female_name2_most
```

```python
top_female_names2 = female_names['Name2'].value_counts(sort=True)
top_female_names2.head(10)
```

## топ по всем  именам
топ по всем мужским именам

```python
all_male_names = top_male_names1.add(top_male_names2, fill_value=0)
top_male_names = all_male_names.sort_values(ascending=False)
top_male_names.head(10)
```

топ по всем женским именам

```python
all_female_names = top_female_names1.add(top_female_names2, fill_value=0)
top_female_names = all_female_names.sort_values(ascending=False)
top_female_names.head(10)
```
