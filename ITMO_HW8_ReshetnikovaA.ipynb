# Получить информацию обо всех книгах и сохранить в pd.DataFrame - 1 балл

hint: используйте метод requests.get()

```python
url = 'https://www.anapioficeandfire.com/api/books'
response = requests.get(url)
books_data = response.json()


books_list = []

for book in books_data:
    books_list.append({
            'name': book['name'],
            'isbn': book['isbn'],
            'number_of_pages': book['numberOfPages'],
            'publisher': book['publisher'],
            'released': book['released'],
            'characters': book['characters']
        })


books_df = pd.DataFrame(books_list)
books_df
```

# Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame - 1 балл

hint: используйте метод requests.get()

```python
url = "https://anapioficeandfire.com/api/houses"
houses = []
def get_houses(url):
    while url:
        response = requests.get(url)
        houses.extend(response.json())
        url = response.links.get('next', {}).get('url')


get_houses(url)

df_houses = pd.DataFrame(houses)
df_houses.head(5)

```

# Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame - 2 балла

hint: используйте метод requests.get(), используйте в ссылке запроса параметр hasTitles для фильтрации

```python
url = "https://anapioficeandfire.com/api/houses?hasTitles=true"
houses_with_titles = []

def get_houses_with_titles(url):
    while url:
        response = requests.get(url)
        houses_with_titles.extend(response.json())
        url = response.links.get('next', {}).get('url')

get_houses_with_titles(url)
df_houses_with_titles = pd.DataFrame(houses_with_titles)

df_houses_with_titles.head(10)
```
